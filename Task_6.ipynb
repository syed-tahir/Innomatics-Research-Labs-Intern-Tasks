Problem Statement 1:Smart Parking Lot Management System

Design a function to manage a smart parking lot.

The system should:

* Accept vehicle entry and exit logs


* Calculate total parked vehicles


* Identify peak parking usage


* Alert if parking exceeds capacity


```python
#Creating the function to Check the availability of parking lot

def check_parking_availability(parking_lot_capacity, vehicle_logs):
    
    vehicle_logs = [log.upper() for log in vehicle_logs]
    vehicle_logs = [log.strip() for log in vehicle_logs]

    vehicles_parked = 0

    for logs in vehicle_logs:
        if logs == "IN":
            vehicles_parked +=1
        elif logs == "OUT":
            vehicles_parked -= 1
    
    if vehicles_parked <= parking_lot_capacity:
        parking_status = "Available"
    else:
        parking_status = "Full"


    return parking_status, vehicles_parked

#Taking the input

parking_lot_capacity = int(input("Parking Capacity: "))
vehicle_logs = input("Vehicle Logs: ").split(",")

#Printing the input

print("Parking Capacity: ", parking_lot_capacity)
print("Vehicle Logs: ", vehicle_logs)


#Calling the function to check the parking availability

parking_status, vehicles_parked = check_parking_availability(parking_lot_capacity, vehicle_logs)


# Output 

print("Currently parked vehicles: ", vehicles_parked)
print("Parking Status: ", parking_status)
```

 Problem Statement 2: Online Food Delivery Time Estimator

Create a function that estimates delivery time based on:

* Distance (km)


* Weather condition


* Traffic level


* Apply delays dynamically and display final ETA.


```python
# Function to calculate delivery time
def time_estimator(distance, traffic_level, weather_condition):
    base_time = distance * 3  # Assuming 3 minutes per kilometer as base time

    # Adjusting time based on traffic level
    if traffic_level == "Low":
        traffic_time = 1
    elif traffic_level == "Medium":
        traffic_time = 1.5
    elif traffic_level == "High":
        traffic_time = 2
    else:
        print("Invalid traffic level. Please enter from Low, Medium, or High.")
        return "Error"
    
    # Adjusting time based on weather condition
    if weather_condition == "Clear":
        weather_time = 1
    elif weather_condition == "Windy":
        weather_time = 1.2
    elif weather_condition == "Rainy":
        weather_time = 1.5
    else:
        print("Invalid weather condition. Please enter from Clear, Windy, or Rainy.")
        return "Error"

    total_time = base_time * traffic_time * weather_time
    return total_time

# Taking input for delivery time estimation
distance = float(input("Enter the distance to the delivery location (in kilometers): "))
traffic_level = input("Enter the traffic level (Low, Medium, High): ")
weather_condition = input("Enter the weather condition (Clear, Rainy, Snowy): ")

# Calculating and printing the estimated delivery time
estimated_time = time_estimator(distance, traffic_level, weather_condition)
if estimated_time != "Error":
    print(f"Estimated delivery time: {estimated_time} minutes")
else:
    print("Error in calculating delivery time. Recheck your inputs.")
```

 Problem Statement 3: Movie Theatre Seat Occupancy Analyzer

Build a function that analyzes seat booking data and:

* Calculates occupancy percentage


* Determines if show is Housefull


* Suggests opening additional shows


```python
# Function to analyze Movie seat occupancy

def analyze_seat_occupancy(total_seats, booked_seats):
    
    # Count booked seats using loop
    booked_count = 0
    for seat in booked_seats:
        if seat == 1:
            booked_count += 1

    # Calculate occupancy percentage
    occupancy = (booked_count / total_seats) * 100

    # Determine show status
    if occupancy == 100:
        status = "Housefull"
    elif occupancy >= 75:
        status = "Almost Full"
    else:
        status = "Available"

    # Suggest additional shows
    if occupancy >= 90:
        suggestion = "Consider opening additional shows."
    else:
        suggestion = "No additional shows needed."

    print("\nMovie Theatre Occupancy Report")
    print(f"Occupancy: {round(occupancy)}%")
    print(f"Show Status: {status}")
    print(f"Suggestion: {suggestion}")


# Sample Input
total_seats = 200
booked_seats = [1] * 150   # 150 booked seats

analyze_seat_occupancy(total_seats, booked_seats)
```

Problem Statement 4:Cloud Server Load Classification System

Create a function to classify server load based on CPU usage readings.

Rules:

* Average CPU < 50% → Normal


* 50%–80% → Warning


* 80% → Critical


```python
# Function to classify server load based on CPU readings

def classify_server_load(cpu_readings):
    total = 0
    
    # Calculate total using loop
    for reading in cpu_readings:
        total += reading
    
    average = total / len(cpu_readings)
    
    # Classification Rules
    if average < 50:
        status = "Normal"
    elif 50 <= average <= 80:
        status = "Warning"
    else:
        status = "Critical"
    
    print("\nCloud Server Load Report")
    print("Average CPU Load:", round(average), "%")
    print("Server Status:", status)


# Sample Input
cpu_readings = [45, 60, 70, 85, 90]

classify_server_load(cpu_readings)
```

Problem Statement 5:Smart Classroom Resource Usage Monitor

Design a function that tracks usage of classroom resources (projector, AC, lights) and identifies overuse patterns.

Real-Time Use

* Smart classrooms


* Energy optimization systems


```python
# Function to monitor resource usage in a smart classroom

def monitor_resource_usage(resource_usage):
    overused = []

    # Define maximum allowed usage (in hours)
    usage_limit = 8

    for resource, hours in resource_usage.items():
        if hours > usage_limit:
            overused.append(resource)

    print("\nSmart Classroom Energy Report")
    
    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")


# Sample Input (User Data)
resource_usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

monitor_resource_usage(resource_usage)
```

Problem Statement 6: Online Event Registration Capacity Controller

Create a function that manages event registrations by:

* Tracking registrations


* Preventing overbooking


* Triggering waitlist mode


```python
#Creating the function to manage event registration

def manage_registration(capacity, total_registrations):
    confirmed = 0
    waitlisted = 0
    
    for i in range(total_registrations):
        if confirmed < capacity:
            confirmed += 1
        else:
            waitlisted += 1
    
    status = "Closed" if confirmed >= capacity else "Open"
    
    print("\nRegistration Summary")
    print("----------------------")
    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)
    print("Registration Status:", status)


# Taking user input
capacity = int(input("Enter Event Capacity: "))
registrations = int(input("Enter Total Registrations: "))

manage_registration(capacity, registrations)
```
