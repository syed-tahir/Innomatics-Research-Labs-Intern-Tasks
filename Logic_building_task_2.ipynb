Problem 1: Unique Words in a Sentence


Problem Statement: 

Given a sentence, identify all unique words and display:

The count of unique words

The unique words themselves

Hint: 

Split the sentence into individual words.

Convert the list of words into a set to remove duplicates.

Print both the set and its length.


```python
sentences = input("Enter a sentence: ")
print("The sentence you entered is: ", sentences)

unique_words = set(sentences.split())

print("The count of unique words is: ", len(unique_words))

print("The unique words are: ", unique_words)
```

Problem 2: Highest Salary from Employee Data

Problem Statement: 

Store employee names and their salaries, and determine which employee has the highest salary.

Hint:

Use a dictionary where the key is the employee name and the value is the salary.

Compare salary values to find the highest.


```python
# Create a dictionary to store employee names and their salaries
employees = {
    "Alice": 50000,
    "Bob": 75000,
    "Charlie": 62000,
    "David": 82000
}

# Find the employee with the highest salary
highest_paid_employee = max(employees, key=employees.get)

# Get the highest salary
highest_salary = employees[highest_paid_employee]

# Display the result
print("Employee Data:", employees)
print("Highest Paid Employee:", highest_paid_employee)
print("Highest Salary:", highest_salary)
```

Problem 3: Find Maximum and Minimum Values

Problem Statement: 

Find the maximum and minimum values from a list of numbers.

Hint:

Initialize both maximum and minimum with the first element of the list.

Use a loop to compare each element.


```python
# List of numbers
numbers = [45, 12, 78, 34, 89, 23, 10]

# Initialize max and min with the first element
maximum = numbers[0]
minimum = numbers[0]

# Loop through the list
for num in numbers:
    if num > maximum:
        maximum = num
    if num < minimum:
        minimum = num

# Display results
print("Numbers:", numbers)
print("Maximum value:", maximum)
print("Minimum value:", minimum)
```

Problem 4: Count Products Above a Price Threshold

Problem Statement:

Count how many products have a price greater than 1000.

Hint:

Use a loop to compare each price with 1000.



```python
# Dictionary storing product names and their prices
products = {
    "Laptop": 55000,
    "Mouse": 500,
    "Keyboard": 1500,
    "Headphones": 1200,
    "USB Cable": 300
}

# Initialize counter
count = 0

# Loop through product prices
for price in products.values():
    if price > 1000:
        count += 1

# Display result
print("Products:", products)
print("Number of products priced above 1000:", count)
```

Problem 5: Calculate Attendance Percentage

Problem Statement:

Calculate the attendance percentage from a list containing "P" (Present) and "A" (Absent).

Hint: 

Count the number of "P" values.

Divide by the total number of days.


```python
# List representing attendance
attendance = ["P", "A", "P", "P", "A", "P", "P"]

# Count number of "P"
present_count = 0

for status in attendance:
    if status == "P":
        present_count += 1

# Total number of days
total_days = len(attendance)

# Calculate percentage
attendance_percentage = (present_count / total_days) * 100

# Display result
print("Attendance Record:", attendance)
print("Present Days:", present_count)
print("Total Days:", total_days)
print("Attendance Percentage:", attendance_percentage, "%")
```

Problem 6: Remove Duplicate Phone Numbers

Problem Statement:

You are given a list of phone numbers that may contain duplicates. Write a Python program to remove duplicate phone numbers and display only the unique phone numbers.

Hint:

Convert the list into a set.

Sets automatically remove duplicate values.


```python
# List containing phone numbers (with duplicates)
phone_numbers = [
    "9876543210",
    "9123456780",
    "9876543210",
    "9988776655",
    "9123456780",
    "9012345678"
]

# Convert list to set to remove duplicates
unique_numbers = set(phone_numbers)

# Display results
print("Original Phone Numbers:", phone_numbers)
print("Unique Phone Numbers:", unique_numbers)
```

Problem 7: Count Character Frequency

Problem Statement:

Count how many times each character appears in a given string.

Hint:

Use a dictionary.

Increment the count for each character occurrence.


```python
# Given string
text = "programming"

# Dictionary to store character frequency
char_count = {}

# Loop through each character
for char in text:
    if char in char_count:
        char_count[char] += 1
    else:
        char_count[char] = 1

# Display result
print("String:", text)
print("Character Frequency:", char_count)
```

Problem 9: Check if a Key Exists in a Dictionary

Problem Statement:

Check whether a specific key exists in a dictionary.

Hint:

Use the in keyword.


```python
# Sample dictionary
student_marks = {
    "Ali": 85,
    "Sara": 92,
    "John": 78,
    "Ayesha": 88
}

# Key to check
key_to_check = "Sara"

# Check if key exists
if key_to_check in student_marks:
    print(f"'{key_to_check}' exists in the dictionary.")
else:
    print(f"'{key_to_check}' does not exist in the dictionary.")
```

Problem 10: Calculate Average Marks

Problem Statement:

Calculate the average marks of a student from a list of marks.

Hint:

Calculate the sum of all values.

Divide by the total number of marks.


```python
# List of marks
marks = [75, 80, 90, 85, 70]

# Calculate total marks
total = sum(marks)

# Calculate number of subjects
count = len(marks)

# Calculate average
average = total / count

# Display result
print("Marks:", marks)
print("Total Marks:", total)
print("Number of Subjects:", count)
print("Average Marks:", average)
```
