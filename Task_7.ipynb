Problem 1: Smart Light Controller

Problem Statement:

Create a class to control a smart light that can be turned ON or OFF and display its status.


```python
# Creating a class for Smart Light
class SmartLight:
    
    # Constructor to initialize light name and default status
    def __init__(self, light_name):
        self.light_name = light_name
        self.status = "OFF"   # Default status
    
    # Method to turn ON the light
    def turn_on(self):
        self.status = "ON"
    
    # Method to turn OFF the light
    def turn_off(self):
        self.status = "OFF"
    
    # Method to display current status
    def display_status(self):
        print(f"{self.light_name} light is {self.status}")


# ðŸ”¹ Taking user input
light_name = input("Enter Light Name: ")
action = input("Enter Action (ON/OFF): ").upper()

# ðŸ”¹ Creating object
light1 = SmartLight(light_name)

# ðŸ”¹ Performing action
if action == "ON":
    light1.turn_on()
elif action == "OFF":
    light1.turn_off()
else:
    print("Invalid Action!")

# ðŸ”¹ Displaying result
light1.display_status()
```

Problem 2: Employee ID Card System

Problem Statement:

Create a class that stores employee name, ID, and department and displays ID card details.


```python
# Creating a class for Employee ID Card
class Employee:
    
    # Constructor to initialize employee details
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department
    
    # Method to display ID card details
    def display_id_card(self):
        print("\nEmployee ID Card")
        print("Name:", self.name)
        print("ID:", self.emp_id)
        print("Department:", self.department)


# ðŸ”¹ Taking user input
name = input("Enter Employee Name: ")
emp_id = input("Enter Employee ID: ")
department = input("Enter Department: ")

# ðŸ”¹ Creating object
employee1 = Employee(name, emp_id, department)

# ðŸ”¹ Displaying ID Card
employee1.display_id_card()
```

Problem 3: Mobile Contact Record

Problem Statement:

Create a class to store contact name and phone number and display contact information.


```python
# Creating a class for Contact
class Contact:
    
    # Constructor to initialize contact details
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone
    
    # Method to display contact information
    def display_contact(self):
        print("\nContact Saved")
        print("Name:", self.name)
        print("Phone:", self.phone)


# ðŸ”¹ Taking user input
name = input("Enter Contact Name: ")
phone = input("Enter Phone Number: ")

# ðŸ”¹ Creating object
contact1 = Contact(name, phone)

# ðŸ”¹ Displaying contact details
contact1.display_contact()
```

Problem 5: Movie Rating Display System

Problem Statement:

Create a class that stores movie name and rating and displays movie details.


```python
# Creating a class for Movie
class Movie:
    
    # Constructor to initialize movie details
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating
    
    # Method to display movie details
    def display_movie_details(self):
        print("\nMovie:", self.name)
        print(f"Rating: {self.rating} / 5")


# ðŸ”¹ Taking user input
movie_name = input("Enter Movie Name: ")
rating = float(input("Enter Rating: "))

# ðŸ”¹ Creating object
movie1 = Movie(movie_name, rating)

# ðŸ”¹ Displaying movie details
movie1.display_movie_details()
```

Problem 6: Delivery Address Manager

Problem Statement:

Create a class that stores customer name and delivery address and prints delivery details.


```python
# Creating a class for Delivery Details
class Delivery:
    
    # Constructor to initialize customer details
    def __init__(self, customer_name, address):
        self.customer_name = customer_name
        self.address = address
    
    # Method to display delivery details
    def display_delivery_details(self):
        print("\nDelivery Details")
        print("Customer:", self.customer_name)
        print("Address:", self.address)


# ðŸ”¹ Taking user input
customer_name = input("Enter Customer Name: ")
address = input("Enter Address: ")

# ðŸ”¹ Creating object
delivery1 = Delivery(customer_name, address)

# ðŸ”¹ Displaying delivery details
delivery1.display_delivery_details()
```
