Problem 1: Social Media – Post Engagement Analyzer

Problem Statement:

Create a function to analyze engagement on a social media post.

Rules:

Calculate total likes

If Total Likes ≥ 1000 → Viral Post

Otherwise → Normal Engagement


```python
# Function to analyze post engagement
def analyze_engagement(likes_list):
    total_likes = 0
    
    # Loop to calculate total likes
    for likes in likes_list:
        total_likes += likes
    
    # Check post status
    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"
    
    # Display result
    print("Total Likes:", total_likes)
    print("Post Status:", status)


# Example usage
likes_data = [200, 150, 300, 250, 150]  # Sample likes from different sources
analyze_engagement(likes_data)
```

Problem 3: Agriculture – Rainfall Adequacy Checker

Problem Statement:


Create a function that analyzes seasonal rainfall data.

Rules:

Calculate average rainfall

Average ≥ required level → Adequate Rainfall

Otherwise → Inadequate Rainfall


```python
# Function to analyze seasonal rainfall
def rainfall_checker(rainfall_data, required_level):
    total_rainfall = 0
    
    # Loop to calculate total rainfall
    for rainfall in rainfall_data:
        total_rainfall += rainfall
    
    # Calculate average rainfall
    average_rainfall = total_rainfall / len(rainfall_data)
    
    # Check rainfall adequacy
    if average_rainfall >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"
    
    # Display result
    print("Average Rainfall:", int(average_rainfall))
    print("Rainfall Status:", status)


# Example usage
seasonal_rainfall = [70, 75, 80, 65, 70]
required_rainfall_level = 70

rainfall_checker(seasonal_rainfall, required_rainfall_level)
```

Problem 4: Social Media – Duplicate Account Detection

Problem Statement:

Create a function to detect duplicate usernames in a system.

Rules:

If duplicates exist → Flag accounts

Else → Accounts are unique


```python
# Function to detect duplicate usernames
def detect_duplicate_accounts(usernames):
    
    # Convert list to set
    unique_usernames = set(usernames)
    
    # Check for duplicates
    if len(unique_usernames) != len(usernames):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")


# Example usage
user_list = ["tahir01", "john23", "alice99", "tahir01", "mark07"]
detect_duplicate_accounts(user_list)
```

Problem 6: Agriculture – Premium Crop Price Filter

Problem Statement:

Create a function that filters crop prices above a premium threshold.

Rules:

Price > ₹2000 → Premium Crop

Else → Regular Crop


```python
# Function to filter premium crop prices
def filter_premium_crops(prices):
    premium_crops = []
    
    # Loop to filter prices above ₹2000
    for price in prices:
        if price > 2000:
            premium_crops.append(price)
    
    # Display result
    print("Premium Crops:", premium_crops)


# Example usage
crop_prices = [1500, 2500, 1800, 3200, 1900]
filter_premium_crops(crop_prices)
```

Problem 9: E-Learning – Student Attendance Eligibility System

Problem Statement:


Create a function to calculate attendance percentage and check exam eligibility.

Rules:

Attendance ≥ 75% → Eligible

Attendance < 75% → Not Eligible


```python
# Function to calculate attendance percentage and check eligibility
def check_attendance(attendance_list):
    total_classes = len(attendance_list)
    attended_classes = 0
    
    # Loop to count attended classes
    for status in attendance_list:
        if status == 1:   # 1 means Present, 0 means Absent
            attended_classes += 1
    
    # Calculate attendance percentage
    attendance_percentage = (attended_classes / total_classes) * 100
    
    # Check eligibility
    if attendance_percentage >= 75:
        eligibility = "Eligible"
    else:
        eligibility = "Not Eligible"
    
    # Display result
    print("Attendance Percentage:", attendance_percentage)
    print("Exam Eligibility:", eligibility)


# Example usage
# 1 = Present, 0 = Absent
attendance_data = [1, 1, 1, 0, 1]   # 4 out of 5 classes attended (80%)
check_attendance(attendance_data)
```

Problem 10: Smart Electricity Bill Analyzer

Problem Statement:

You are building a smart electricity billing system for residential users.

Create a function that calculates the electricity bill and classifies usage level based on units consumed.

Billing Rules:

First 100 units → ₹3 per unit

Next 100 units (101–200) → ₹5 per unit

Above 200 units → ₹7 per unit

Usage Classification:

Total bill < ₹500 → Low Usage

₹500 – ₹1500 → Moderate Usage

₹1500 → High Usage



```python
# Function to calculate electricity bill and classify usage
def calculate_electricity_bill(units):
    bill = 0
    
    # Billing slab calculation
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)
    
    # Usage classification
    if bill < 500:
        usage_status = "Low Usage"
    elif 500 <= bill <= 1500:
        usage_status = "Moderate Usage"
    else:
        usage_status = "High Usage"
    
    return bill, usage_status


# Example usage
units_consumed = 250
total_bill, status = calculate_electricity_bill(units_consumed)

print("Units Consumed:", units_consumed)
print("Total Bill: ₹", total_bill)
print("Usage Level:", status)
```
